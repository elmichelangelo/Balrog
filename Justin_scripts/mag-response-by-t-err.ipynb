```python
import numpy as np
import fitsio
from astropy.table import Table, join
import matplotlib.pyplot as plt
from matplotlib import colors
import os

import seaborn as sb
plt.style.use('seaborn')
sb.set_context("notebook", font_scale=1.5)
```

```python
class MidpointNormalize(colors.Normalize):
    """
    Normalize the colorbar so that diverging bars work from either
    side of a set midpoint value
    """
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y), np.isnan(value))
```

```python
run = 'y3-merged'
version = 1.2

bal_dir = f'/project/projectdirs/des/severett/Balrog/{run}/stacked_catalogs/{version}/sof/'

match_file = os.path.join(bal_dir, f'balrog_matched_catalog_sof_{run}_v{version}.fits')
det_file   = os.path.join(bal_dir, f'balrog_detection_catalog_sof_{run}_v{version}.fits')

# Desired columns from matched catalog
match_cols = ['bal_id', 'meas_cm_mag_deredden', 'true_bdf_mag_deredden',
              'meas_cm_T', 'true_bdf_T', 'true_bdf_T_err']

# Desired columns from detection catalog
det_cols = ['bal_id', 'meas_FLAGS_GOLD_SOF_ONLY', 'match_flag_1.5_asec', 'meas_EXTENDED_CLASS_SOF',
            'flags_foreground', 'flags_footprint', 'flags_badregions']

print('Loading matched file...')
match = Table(fitsio.read(match_file, columns=match_cols))
print('Loading det file...')
det = Table(fitsio.read(det_file, columns=det_cols))
print('joining...')
bal = join(match, det, keys='bal_id', join_type='inner')

# Make desired cuts (Standard Y3 GOLD cuts in this case)
cuts = np.where(
    (bal['meas_FLAGS_GOLD_SOF_ONLY'] < 2) &
    (bal['flags_foreground'] == 0 ) &
    (bal['flags_footprint'] == 1) &
    (bal['flags_badregions'] < 2) &
    (bal['meas_EXTENDED_CLASS_SOF'] > 1) &
    (bal['match_flag_1.5_asec'] < 2)
)

print('Making cuts...')
bal = bal[cuts]
```

```python
sb.set_style("whitegrid")
```

```python
xlim = [15, 26]
ylim = [-8, 4]

vmin, vmax = -2.5, 5

bindx = dict(zip('griz', range(4)))
b = 'i'
bi = bindx[b]

print('Plotting sample...')

N = 10 # Only plot 10% to run faster

T_err = bal['meas_cm_T'] - bal['true_bdf_T']
T_err = np.log10(np.abs(T_err))

x = bal['true_bdf_mag_deredden'][::N,bi]
y = bal['meas_cm_mag_deredden'][::N,bi] - x
    
cmap = 'coolwarm'
plt.scatter(x, y, s=7, c=T_err[::N], cmap=cmap,
            norm=MidpointNormalize(midpoint=0, vmin=vmin, vmax=vmax))

# Reference lines
plt.axhline(-1, lw=3, ls=':', c='k')
plt.axhline(0, lw=3, ls='--', c='k')
plt.axhline(1, lw=3, ls=':', c='k')

plt.xlim(xlim)
plt.ylim(ylim)

plt.xlabel('True {}-mag (bdf)'.format(b))
plt.ylabel('Measured-True {}-mag (cm-bdf)'.format(b))
cbar = plt.colorbar()
cbar.set_label('log10(|Meas-True T|)', rotation=270, labelpad=25)
  
```
